### Step 3: Data Cleaning and Transformation

This step is critical for ensuring our analysis is accurate. We will handle missing values and correct data types.

#### **Theoretical Concept: Data Type Conversion & Handling Nulls**
Data often comes in non-ideal formats. Storing dates as strings, for example, prevents us from extracting components like the year or month, or from plotting data over time. Converting columns to their proper data types (`pd.to_datetime`, `.astype()`) is a fundamental preprocessing step.

For null values, we have several strategies:
1.  **Drop:** If only a very small percentage of rows have missing data, dropping them might be acceptable (`.dropna()`).
2.  **Fill/Impute:** Replace missing values with a placeholder (like "Unknown") or a statistical measure (like the mode for categorical data). This is useful when you don't want to lose the other information in those rows.

```python
# 1. Handle missing values in 'director' and 'cast'
# Since these are text fields and many are missing, we'll fill them with 'Unknown'.
netflix_df['director'] = netflix_df['director'].fillna('Unknown')
netflix_df['cast'] = netflix_df['cast'].fillna('Unknown')
```

```python
# 2. Handle missing 'country'
# We'll fill with the mode, which is the most common country.
mode_country = netflix_df['country'].mode()[0]
netflix_df['country'] = netflix_df['country'].fillna(mode_country)
```

```python
# 3. Drop the few rows with missing 'date_added' and 'rating'
# Since the number is small (less than 0.2% of data), dropping them is a safe option.
netflix_df.dropna(subset=['date_added', 'rating'], inplace=True)
```

```python
# 4. Convert 'date_added' to datetime objects
# Use format='mixed' to handle potential variations in date formats
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'], format='mixed', dayfirst=False)
```

* **format='mixed':** This argument tells pandas to infer the date format automatically. This is helpful when the date strings in the column have different formats.

* **dayfirst=False:** This argument specifies that when the date format is ambiguous (e.g., 01/02/2023), it should be interpreted as month first (January 2nd) rather than day first (February 1st).

```python
# 5. Create new features for year and month added
netflix_df['year_added'] = netflix_df['date_added'].dt.year
netflix_df['month_added'] = netflix_df['date_added'].dt.month
```

```python
# Verify our cleaning and transformation
print("Missing values after cleaning:")
print(netflix_df.isnull().sum())
print("\nData types after transformation:")
print(netflix_df.dtypes)
```

#### 4.3 What are the most popular genres?

#### **Theoretical Concept: Handling Multi-Value Text Columns**
The `listed_in` column contains strings with multiple genres separated by commas (e.g., "Dramas, International Movies"). To analyze each genre individually, we need to transform the data. A common technique is to:
1.  **Split** the string in each row into a list of genres.
2.  **Explode** the DataFrame so that each genre in the list gets its own row, duplicating the other information for that title.
This allows us to perform a `value_counts()` on the genres.

```python
# Split the 'listed_in' column and explode it
genres = netflix_df.assign(genre=netflix_df['listed_in'].str.split(', ')).explode('genre')
```

```python
# Get the top 15 genres and their counts
top_genres_counts = genres['genre'].value_counts().reset_index()
top_genres_counts.columns = ['genre', 'count'] # Rename columns for clarity

# Select only the top 15 for plotting
top_genres_counts_plot = top_genres_counts.head(15)

plt.figure(figsize=(12, 8))
sns.barplot(y='genre', x='count', data=top_genres_counts_plot, palette='mako', hue='genre', legend=False)
plt.title('Top 15 Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

**Insight:** "International Movies" is the most common genre tag, highlighting Netflix's global content strategy. This is followed by Dramas, Comedies, and Action & Adventure.

Let's create a new feature to analyze how old content is when it gets added to Netflix. This can tell us about their acquisition strategy (buying old classics vs. releasing new originals).

lets examine how has the distribution of content ratings changed over time?

```python
# Approach:
# - Aggregate counts of titles by rating over time (year_added or release_year)
# - Show stacked area chart or proportion per year.
# ----------------------------

# We'll use year_added (when it was added to Netflix) as the timeline for the distribution change.
# Filter to rows that have a year_added and rating
q1 = netflix_df.dropna(subset=['year_added', 'rating']).copy()
# Group by year_added and rating
rating_year = q1.groupby(['year_added', 'rating']).size().unstack(fill_value=0)

# Optional: keep only years with reasonable data (e.g., 2008+)
rating_year = rating_year.loc[rating_year.index >= 2008]

# Normalize to proportions for stacked area (showing share)
rating_prop = rating_year.div(rating_year.sum(axis=1), axis=0)

# Plot: stacked area of proportions (so we can see how rating mix changes)
rating_prop.plot.area(figsize=(12,7), alpha=0.85)
plt.title("Distribution of Content Ratings Over Time (by Year Added)")
plt.ylabel("Proportion of Titles")
plt.xlabel("Year Added to Netflix")
plt.legend(title="Rating", bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()


# Also show absolute counts (stacked) if useful
rating_year.plot(kind='bar', stacked=True, figsize=(12,7))
plt.title("Counts of Titles by Rating and Year Added")
plt.ylabel("Number of Titles")
plt.xlabel("Year Added")
plt.tight_layout()
plt.show()

# Quick summary table: top rating for each year
top_rating_each_year = rating_year.idxmax(axis=1)
top_rating_each_year.head(10)
```

Insight:
Content additions surged sharply from 2016 to 2020, with 2019 being the peak year. TV-14, TV-MA, and TV-PG dominate the catalog growth, showing Netflix‚Äôs focus on teen and mature audiences. The consistent rise across ratings also reflects a broad strategy rather than targeting a single audience segment.

 Is there a relationship between content age and its type (Movie vs. TV Show)?
- Approach:
    - Content age = (year_added - release_year) or use release_year distribution by type.
    - Compare age distributions between 'Movie' and 'TV Show'

```python
# Compute 'age_at_add' = year_added - release_year (how old the content was when Netflix added it)
age_df = netflix_df.dropna(subset=['year_added', 'release_year', 'type']).copy()
age_df['age_at_add'] = age_df['year_added'] - age_df['release_year']

# Remove extreme outliers or implausible negative ages where release_year > year_added:
# Keep rows where age_at_add is reasonable (e.g., -1 to 100). Keep negative values too if they occur but we can examine them separately.
age_df = age_df[(age_df['age_at_add'].notna()) & (age_df['age_at_add'] > -5) & (age_df['age_at_add'] < 200)]

# Boxplot of age_at_add by type
plt.figure(figsize=(9,6))
sns.boxplot(data=age_df, x='type', y='age_at_add')
plt.title("Age of Content at Time Added to Netflix by Type")
plt.ylabel("Years between release and being added to Netflix")
plt.xlabel("Type")
plt.ylim(-2, 40)  # zoom in to show central tendencies
plt.show()
```

```python
# Show descriptive stats
age_stats = age_df.groupby('type')['age_at_add'].describe()[['count','mean','50%','std','min','max']]
age_stats.columns = ['count','mean','median','std','min','max']
print(age_stats)

# Additional: histogram / KDEs
plt.figure(figsize=(10,6))
sns.kdeplot(data=age_df, x='age_at_add', hue='type', common_norm=False)
plt.title("Density of Age at Add by Type")
plt.xlim(-2, 40)
plt.show()
```

Insight:
Movies are typically added to Netflix very shortly after their release (median age ~0‚Äì1 years), while TV shows tend to be older when added. TV shows show a wider spread and many long-tail outliers (15‚Äì40+ years), meaning Netflix often acquires older or classic series. Overall, Netflix adds movies more quickly, while TV shows include both recent and much older content.

Find trends in content production based on release_year vs year_added
- Approach:
     - Scatter or 2D histogram of release_year vs year_added
     - Produce 'lag' distribution = year_added - release_year; plot median lag by release_year or by added_year

```python
import datetime

# Clean data for plotting
trend_df = netflix_df.dropna(subset=['release_year', 'year_added']).copy()
trend_df = trend_df[(trend_df['release_year'] >= 1900) & (trend_df['release_year'] <= datetime.datetime.now().year)]

# Scatter: release_year vs year_added (if year_added is missing for many rows this will drop them)
plt.figure(figsize=(10,8))
plt.scatter(trend_df['release_year'], trend_df['year_added'], alpha=0.4, s=15)
plt.xlabel("Release Year")
plt.ylabel("Year Added to Netflix")
plt.title("Release Year vs Year Added (each point = a title)")
plt.xlim(trend_df['release_year'].min()-1, trend_df['release_year'].max()+1)
plt.ylim(trend_df['year_added'].min()-1, trend_df['year_added'].max()+1)
plt.show()

# 2D density: helps show trends
plt.figure(figsize=(10,8))
sns.kdeplot(x=trend_df['release_year'], y=trend_df['year_added'], cmap="Blues", fill=True, thresh=0.05)
plt.title("Release Year vs Year Added (2D density)")
plt.xlabel("Release Year")
plt.ylabel("Year Added")
plt.show()

# Lag analysis
trend_df['lag_years'] = trend_df['year_added'] - trend_df['release_year']
# Median lag by release_year (for years with enough titles)
median_lag_by_release = trend_df.groupby('release_year')['lag_years'].median().dropna()
# Plot (restrict to release years with at least N titles to reduce noise)
counts_by_release = trend_df.groupby('release_year').size()
valid_years = counts_by_release[counts_by_release >= 10].index
plt.figure(figsize=(12,6))
plt.plot(median_lag_by_release.loc[valid_years])
plt.title("Median Lag (Year Added - Release Year) by Release Year (only release years with >=10 titles)")
plt.xlabel("Release Year")
plt.ylabel("Median Lag (years)")
plt.show()

# Distribution of lag overall
plt.figure(figsize=(10,5))
sns.histplot(trend_df['lag_years'].clip(-5,80), bins=50, kde=True)
plt.title("Distribution of Lag between Release Year and Year Added")
plt.xlabel("Lag (years)")
plt.xlim(-3,50)
plt.show()

# Quick insights to compute:
print("Mean lag:", trend_df['lag_years'].mean())
print("Median lag:", trend_df['lag_years'].median())
print("Percent added within 1 year of release:", (trend_df['lag_years'] <= 1).mean())
```

Find most common word pairs or phrases in content descriptions
 - Use CountVectorizer with ngram_range=(2,2) (bigrams) on the 'description' column
 - Preprocessing: lowercase, remove punctuation (CountVectorizer can handle token pattern)
 - Return top N bigrams

```python
from sklearn.feature_extraction.text import CountVectorizer

descriptions = netflix_df['description'].dropna().astype(str)

# A simple preprocessor/cleaner for the vectorizer: remove extra whitespace
def simple_clean(text):
    return text.lower()

vectorizer = CountVectorizer(ngram_range=(2,2), stop_words='english', token_pattern=r"(?u)\b\w+\b", min_df=5)
X = vectorizer.fit_transform(descriptions.apply(simple_clean))

# Sum counts and get top bigrams
bigrams_counts = np.array(X.sum(axis=0)).ravel()
bigrams = vectorizer.get_feature_names_out()
top_n = 50
top_idx = np.argsort(bigrams_counts)[::-1][:top_n]
top_bigrams = [(bigrams[i], bigrams_counts[i]) for i in top_idx]

print("Top bigrams in descriptions (phrase, count):")
for phrase, cnt in top_bigrams[:30]:
    print(phrase, cnt)

# If you want trigrams as well:
vectorizer_tri = CountVectorizer(ngram_range=(3,3), stop_words='english', token_pattern=r"(?u)\b\w+\b", min_df=3)
Xtri = vectorizer_tri.fit_transform(descriptions.apply(simple_clean))
tri_counts = np.array(Xtri.sum(axis=0)).ravel()
tri_names = vectorizer_tri.get_feature_names_out()
top_tri = sorted(zip(tri_names, tri_counts), key=lambda x: x[1], reverse=True)[:20]
print("\nTop trigrams:")
for t,c in top_tri:
    print(t,c)

# Optionally, show the top bigrams per 'type' (Movie vs TV Show)
for t in ['Movie','TV Show']:
    desc_t = netflix_df[netflix_df['type']==t]['description'].dropna().astype(str)
    vec = CountVectorizer(ngram_range=(2,2), stop_words='english', min_df=3)
    X_t = vec.fit_transform(desc_t.apply(simple_clean))
    counts_t = np.array(X_t.sum(axis=0)).ravel()
    names_t = vec.get_feature_names_out()
    top = sorted(zip(names_t, counts_t), key=lambda x: x[1], reverse=True)[:10]
    print(f"\nTop bigrams for {t}:")
    for b,c in top:
        print(b,c)
```
